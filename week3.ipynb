# First Plot with Matplotlib

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.show()
```

# Adding Labels and Titles

```python
plt.plot(x, y)
plt.title("Simple Line Chart")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

# Bar Chart Example

```python
categories = ["Food", "Transport", "Entertainment"]
amounts = [120, 50, 80]

plt.bar(categories, amounts)
plt.title("Monthly Expenses")
plt.show()
```

# Using Seaborn for Quick Styling

```python
import seaborn as sns

sns.barplot(x=categories, y=amounts)
plt.title("Monthly Expenses")
plt.show()
```

# Real Data Example

```python
import pandas as pd

data = pd.read_csv("expenses.csv")
sns.barplot(x="Category", y="Amount", data=data)
plt.title("Daily Expenses by Category")
plt.show()
```
